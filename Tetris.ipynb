```python
def check_fitting(arr, piece, x, y, w, h):
    for i in range(4):
        print(x-piece[i][1], y+piece[i][0])
        if x-piece[i][1] >= h or x-piece[i][1] < 0 or y+piece[i][0] >= w or y+piece[i][0] < 0:
            return False
        elif arr[x-piece[i][1]][y+piece[i][0]] != 0:
            return False
    
    return True

def place_piece(arr, piece, x, y, w, h):
    for i in range(4):
        arr[x-piece[i][1]][y+piece[i][0]] = 1

    for i in range(w):
        for j in reversed(range(h)):
            if arr[j][i] == 0:
                new_xy = [i, j]
                return new_xy

def unplace_piece(arr, piece, x, y):
    for i in range(4):
        arr[x-piece[i][1]][y+piece[i][0]] = 0
    
def helper_tetris(arr, pieces, x, y, w, h):       
    for key, value in pieces.items():
        if value[1] == False and check_fitting(arr, value[0], x, y, w, h):
            new_xy = place_piece(arr, value[0], x, y, w, h)
            pieces[key][1] = True
            print(value[0])
            print(arr)
            print(new_xy)
            if new_xy is None:
                return True
            if helper_tetris(arr, pieces, new_xy[1], new_xy[0], w, h):
                return True
            else:
                pieces[key][1] = False
                unplace_piece(arr, value[0], x, y)
    return False
        
def pack_tetris(w, h, pieces):
    # w - width, h - height, tetris pieces
    
    arr = [[0 for i in range(0, w)] for i in range(0, h)]
    pieces_map = {}
    for piece in range(len(pieces)):
        pieces_map[piece] = [pieces[piece], False]

    print(helper_tetris(arr, pieces_map, h-1, 0, w, h))
```
